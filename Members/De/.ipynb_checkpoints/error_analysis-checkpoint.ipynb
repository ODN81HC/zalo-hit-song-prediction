```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("results.csv")
df1 = df[df["mae"] > 5]
df2 = df[df["mae"] <= 3]
df["big_err"] = df.mae > 3

ax = df.numsongInAlbum.hist(by=df["big_err"],normed=True, sharey=True,bins=34)
ax[0].set_title("NumsongInAlbum(Small Error)")
ax[1].set_title("NumsongInAlbum(Big Error)")
```

```python
# df_err = df.sort_values(by="mae", ascending=False)
df_err= df.sort_values(by="artist_id")
df_err[["ID","label", "prediction", "mae","artist_id", "numsongInAlbum","album_tmp", "track","albumHashAndNameAndReleaseday"]].head(1000)
```

```python
# df.groupby("no_artist").mae.median().plot(style=".")
# sns.scatterplot(x="freq_artist", y="mae", data=df, alpha=0.2)
df1["mean_score_by_artist"] = df1.groupby("artist_id_min_cat").label.transform("median")
df1["median_score_by_artist"] = df1.groupby("artist_id_min_cat").label.transform("mean")
df1["min_score_by_artist"] = df1.groupby("artist_id_min_cat").label.transform("min")
df1["count_score_by_artist"] = df1.groupby("artist_id_min_cat").label.transform("count")



df1[["ID","mean_score_by_artist","median_score_by_artist","min_score_by_artist", "label"]]
```

```python
title_features = ["islyric","isRemix", "isOST", "isBeat", "isVersion", "isCover", "isRemixAlbum", "isOSTAlbum", "isSingleAlbum", "album_name_is_title_name",
                  "isBeatAlbum", "isCoverAlbum", "isSingleAlbum"]

fig, axes = plt.subplots(11, 2, figsize=(20, 55))
for feature, ax in zip(title_features, axes):
    if feature =="islyric":
        print(df1[df1.islyric==True][["ID", "title","artist_id" ,"label"]])
    sns.violinplot(x=feature, y="label", data=df1, inner="quartile",ax=ax[0])
    sns.violinplot(x=feature, y="label", data=df2, inner="quartile",ax=ax[1])    
    ax[0].set_ylim([-2,13])
    ax[1].set_ylim([-2,13])    
```
